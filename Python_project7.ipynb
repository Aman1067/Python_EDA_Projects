```python
df.dropna(axis=0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Rank']=df['Rank'].str[0:-2]
```

```python
df['Rank']=df['Rank'].str.replace(',','').astype('int')
```

```python
df.info()
```

```python
df['Video Uploads']=df['Video Uploads'].astype('int')
```

```python
df['Subscribers']=df['Subscribers'].astype('int')
```

```python
# Replace '-- ' with NaN and then convert to integer
df['Subscribers'] = pd.to_numeric(df['Subscribers'].replace('-- ', np.nan), errors='coerce').astype('Int64')
```

```python
df['Grade'].unique()
```

```python
df['Grade']=df['Grade'].map({'A++ ':5,'A+ ':4,'A ':3,'A- ':2,'B+ ':1})
df.head()
```
